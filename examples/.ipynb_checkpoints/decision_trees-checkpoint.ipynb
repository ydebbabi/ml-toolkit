```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor, plot_tree

X, y = load_boston(return_X_y=True)

regressor = DecisionTreeRegressor(
    criterion="mse",
    max_depth=2, 
    min_samples_leaf=10,
).fit(X, y)

plot_tree(regressor)
plt.show()
```
